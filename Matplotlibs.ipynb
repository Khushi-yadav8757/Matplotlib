```python
# plt.plot(x, y, color='color name', linestyle='line style', marker='marker symbol', label='label name')

import matplotlib.pyplot as plt  

months = [1, 2, 3, 4]           # x-axis data (months)
sales = [1000, 1500, 1200, 1800]  # y-axis data (sales per month)

# create a line plot with custom color, dashed line, circle markers, and label
plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o', label='2025 Sales Data')

plt.xlabel('Months')                   # label x-axis
plt.ylabel('Sales per Month')          # label y-axis
plt.title('Monthly Sales Data Report') # add chart title
plt.legend()                           # show legend
plt.show()                             # display line chart
```

 # BAR CHART

```python
import matplotlib.pyplot as plt  

vehicles = ['Car', 'Bikes', 'Cycle', 'Bus']   # vehicle categories
sales = [1000, 1500, 800, 1200]              # sales data for each vehicle

plt.bar(vehicles, sales, color='orange', label='Vehicles Sales 2025')  # create bar chart
plt.xlabel('Vehicles')             # label x-axis
plt.ylabel('Sales')                # label y-axis
plt.title('Vehicles Sales Analysis')  # add chart title
plt.legend()                       # show legend
plt.show()                         # display bar chart
```

# Pie Chart


```python
import matplotlib.pyplot as plt  

region = ['north', 'south', 'east', 'west']   # region names (labels)
revenue = [3000, 2000, 1500, 1000]            # revenue data for each region

# create pie chart with percentage labels and custom colors
plt.pie(revenue, labels=region, autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightblue', 'coral'])

plt.title("Revenue Contribution By Region")   # add chart title
plt.show()                                   # display pie chart
```

# Histogram

```python
import matplotlib.pyplot as plt  

scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]  # students' scores data

plt.hist(scores, bins=5, color='red', edgecolor='black')  # create histogram with 5 bins
plt.xlabel('Score Range')          # label x-axis
plt.ylabel('Number of Students')   # label y-axis
plt.title('Score Distribution of Students')  # add chart title
plt.show()                         # display histogram
```

# Scatter Plot

```python

import matplotlib.pyplot as plt  

hour_studies = [1,2,3,4,5,6,7,8,9]     # x-axis data (hours studied)
exams_scores = [50,55,60,65,70,75,80,85,90]  # y-axis data (exam scores)

plt.scatter(hour_studies, exams_scores, color='green', marker='o', label='Student Data')  # create scatter plot
plt.xlabel('Hours Studied')            # label x-axis
plt.ylabel('Exam Scores')              # label y-axis
plt.title('Relationship Between Study Time & Exam Score')  # add chart title
plt.legend()                           # show legend
plt.grid(True)                         # show gridlines
plt.show()                             # display plot
```

# SubPlot


```python
# Import the matplotlib library for plotting
import matplotlib.pyplot as plt

# Define data for x and y axes
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

# Create the first subplot (1 row, 2 columns, position 1)
# This means we are dividing the figure into 1 row and 2 columns,
# and the current plot will be in the first position.
plt.subplot(1, 2, 1)
plt.plot(x, y)             # Plot a line chart using x and y
plt.title('Line Chart')    # Add a title to the line chart

# Create the second subplot (1 row, 2 columns, position 2)
# This will place the bar chart next to the line chart.
plt.subplot(1, 2, 2)
plt.bar(x, y)              # Plot a bar chart using x and y
plt.title('Bar Chart')     # Add a title to the bar chart

# Display both subplots on the same figure window
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with 1 row, 2 columns
fig, ax = plt.subplots(1, 2, figsize=(10,8))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

# Plot line chart
ax[0].plot(x, y, color='blue')
ax[0].set_title('Line plot')

# Plot bar chart
ax[1].bar(x, y, color='green')
ax[1].set_title('Bar Chart')

# Add a main title for the figure
fig.suptitle('Comparison of Line and Bar Chart')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
